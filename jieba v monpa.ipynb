```python
# -*- coding: utf-8 -*-
import monpa
import csv
import jieba
import sys
sys.path.append('/Users/NRB/Documents/lawsnote/build_dictionary')
import word_compare

def LongCut(long_sentence, split_char):
  seg = []
  for item in long_sentence.split(split_char):
    if item != "\n": seg.extend(monpa.cut(str(item+split_char)))
  return seg[:-1]

t=input('請輸入欲斷詞的文字:')
```

```python
#set monpa
monpa.load_userdict("/Users/NRB/Documents/lawsnote/build_dictionary/mydict_monpa.txt")
list1=LongCut(t,'。')



#set jieba 
f = open('/Users/NRB/Documents/lawsnote/build_dictionary/mydict_final.txt',encoding='UTF-8')
jieba.set_dictionary("/Users/NRB/Documents/lawsnote/dict.txt.big")
jieba.load_userdict(f)
f.close()

list2 = jieba.lcut(t,cut_all=False,HMM=False)
```

```python
try:
    word_compare.word_compare(list1,list2)
except:
    print("monpa:",list1)
    print("jieba:",list2)
```

兩邊的user dictionary是一樣的，monpa還是有亂斷詞的時候  ex:判處有期徒刑->判，處有期徒刑。
雖然jieba也常常斷錯 ex: 據上論結->據、上、論、結
