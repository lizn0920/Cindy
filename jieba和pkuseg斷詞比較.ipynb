# jieba和pkuseg斷詞比較

在訓練過pkuseg的model之後（訓練檔請見上一篇），比較兩個斷詞套件在「法院判決」上的效果

```python
# -*- coding: utf-8 -*-
import pkuseg
import csv
import jieba
import sys
sys.path.append('/Users/NRB/Documents/lawsnote/build_dictionary')
import word_compare
```

```python
seg = pkuseg.pkuseg(model_name='/Users/NRB/Documents/lawsnote/wordAnalysis/pkuseg',user_dict='/Users/NRB/Documents/lawsnote/build_dictionary/mydict_final.txt')

t=input('請輸入欲斷詞的文字:')

list1=seg.cut(t)

with open('pk3.csv', 'w', newline='',encoding='utf-8') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(list1)

f = open('/Users/NRB/Documents/lawsnote/build_dictionary/mydict_final.txt',encoding='UTF-8')
jieba.set_dictionary("/Users/NRB/Documents/lawsnote/dict.txt.big")
jieba.load_userdict(f)

list2 = jieba.lcut(t,cut_all=False,HMM=False)

f.close()

with open('jb3.csv', 'w', newline='',encoding='UTF-8') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(list2)
```

    word_compare是我寫的小程式，只要遇到斷詞不同處就會輸出詞跟詞的index，但當詞斷得太不一樣時就會有bug。
    簡單來說就是比較兩者不同之處（前：pkuseg，後：jieba）
    可以見到在斷詞上，pkuseg容易把不同的詞斷在一起，jieba在斷數字和人名時雖然沒那麼精準，不過數字人名不是我們的重點，所以沒關係

```python
try:
    word_compare.word_compare(list1,list2)
except:
    print("pkseg:",list1)
    print("jieba:",list2)
```
